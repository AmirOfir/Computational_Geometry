```python
def area(x1, y1, x2, y2, x3, y3): 
    return abs((x1 * (y2 - y3) + x2 * (y3 - y1) + x3 * (y1 - y2)) / 2.0) 
  
def findEnvelope(arrPoints):
    (minX, minY) = (9999,9999)
    (maxX, maxY) = (-9999,-9999)
    for p in arrPoints:
        if p[0] < minX:
            minX = p[0]
        if p[1] < minY:
            minY = p[1]
        if p[0] > maxX:
            maxX = p[0]
        if p[1] > maxY:
            maxY = p[1]
    return minX,minY,maxX,maxY
''' construct a right-angled triange. '''
def constructCircumscribedTriangle(min_x,min_y,max_x,max_y):
    #padding
    p0x = min_x - 10
    p0y = min_y - 10
    p1x = max_x + 10
    p1y = max_y + 10
    #the hypotenuse y=-x+b. passes through (p1x,p1y), x=p0x, y=p0y
    b=p1x+p1y
    upper_point_y = -p0x+b
    right_point_x = b-p0y
    return [[p0x,p0y], [p0x,upper_point_y], [right_point_x, p0y]]
  
class Circle:
    def __init__(self,p1,p2,p3):
        """
        Returns the center and radius of the circle passing the given 3 points.
        In case the 3 points form a line, returns (None, infinity).
        """
        temp = p2[0] * p2[0] + p2[1] * p2[1]
        bc = (p1[0] * p1[0] + p1[1] * p1[1] - temp) / 2
        cd = (temp - p3[0] * p3[0] - p3[1] * p3[1]) / 2
        det = (p1[0] - p2[0]) * (p2[1] - p3[1]) - (p2[0] - p3[0]) * (p1[1] - p2[1])

        if abs(det) < 1.0e-6:
            return (None, np.inf)

        # Center of circle
        self.cx = (bc*(p2[1] - p3[1]) - cd*(p1[1] - p2[1])) / det
        self.cy = ((p1[0] - p2[0]) * cd - (p2[0] - p3[0]) * bc) / det
        self.radius = np.sqrt((cx - p1[0])**2 + (cy - p1[1])**2)
    def containsPoint(self,p):
        dis = ((p[0]-self.cx)**2+(p[1]-self.cy)**2)**0.5
        return dis <= self.radius
class Triangle:
    def __init__(self,p0,p1,p2):
        self.p0=p0
        self.p1=p1
        self.p2=p2
        
    # A function to check whether point P(x, y) 
    # lies inside the triangle    
    def containsPoint(self,p):
        x1=self.p0[0]
        y1=self.p0[1]
        x2=self.p1[0]
        y2=self.p1[1]
        x3=self.p2[0]
        y3=self.p2[1]
        x=p[0]
        y=p[1]
        # Calculate area of triangle ABC 
        A = area (x1, y1, x2, y2, x3, y3) 
        # Calculate area of triangle PBC  
        A1 = area (x, y, x2, y2, x3, y3) 
        # Calculate area of triangle PAC  
        A2 = area (x1, y1, x, y, x3, y3) 
        # Calculate area of triangle PAB  
        A3 = area (x1, y1, x2, y2, x, y) 
        # Check if sum of A1, A2 and A3 is same as A 
        return A == A1 + A2 + A3
class HistoryDAGItem:
    def __init__(self,value, parent):
        self.pointsTo=[]
        self.value=value
        self.parent=parent
    def findChild(self, p):
        for i in self.pointsTo:
            if (i.value.containsPoint(p)):
                return i
        return None
    def append(self, triangles):
        self.pointsTo = [HistoryDAGItem(i,self) for i in triangles]
class HistoryDAG:
    def __init__(self, triangle):
        self.root = HistoryDAGItem(triangle, None)
    def find(self,p):
        curr=self.root
        nextItem = curr.findChild(p)
        while (not (nextItem is None)):
            curr = nextItem
            nextItem = curr.findChild(p)
        return curr
    def _subItems(self, arr, item):
        arr.append(item.value)
        for p in item.pointsTo:
            self._subItems(arr, p)
    def allValues(self):
        a = []
        self._subItems(a, self.root)
        return a
    
points = np.array([[5, 30], [10, -11], [30, 1], [40, 40], [60,20]])
min_x,min_y,max_x,max_y = findEnvelope(points)
root_triangle_points = constructCircumscribedTriangle(min_x,min_y,max_x,max_y)
root_triangle = Triangle(root_triangle_points[0],root_triangle_points[1],root_triangle_points[2])
history_dag = HistoryDAG(root_triangle)


def test_edge_flip():
    pass
for index in range(len(points)):
    p = points[index]
    container = history_dag.find(p)
    container_triangle = container.value
    new_triangles = [Triangle(container_triangle.p0, container_triangle.p1, p), 
                    Triangle(container_triangle.p1, container_triangle.p2, p),
                    Triangle(container_triangle.p2, container_triangle.p0, p)]
    container.append(new_triangles)










import matplotlib.pyplot as plt
triangles = history_dag.allValues()
for triangle in triangles:
    triangle_points = np.array([triangle.p0, triangle.p1, triangle.p2])
    plt.triplot(triangle_points[:,0], triangle_points[:,1], [[0,1,2]])
plt.plot(points[:,0], points[:,1], 'o')
plt.show()
```
